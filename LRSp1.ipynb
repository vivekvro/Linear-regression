```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
df = pd.read_csv('insurance_data.csv')
sns.lmplot(x='Age',y='Premium',data=df)
```

```python
reg = linear_model.LinearRegression()
```

```python
x= df[['Age']]
y=df[['Premium']]
reg.fit(x,y)
```

```python
reg.coef_
reg.intercept_
reg.predict([[25]])
```

```python
def linear_equation():
    slope = reg.coef_
    intercept = reg.intercept_
    indi_var = int(input("Enter indi variable: "))  
    Depend_var = slope * indi_var + intercept
    return Depend_var

linear_equation()
```
